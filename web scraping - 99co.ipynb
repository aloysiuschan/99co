```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup

from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
### connect to the webpage
driver = webdriver.Chrome("/Users/aloysiuschan/Desktop/chromedriver")
driver.get("https://www.99.co/singapore/sale/hdb")
```

```python
### initialize lists
address_list = []
rooms_list = []
year_list = []
price_list = []
bedrooms_list = []
bathrooms_list = []
pricepsf_list = []
sqf_list = []
```

```python
### loop over all the pages, and extract information

pagenum = 1

while True:
    
    my_page = driver.page_source
    my_html = BeautifulSoup(my_page, "html.parser") # get html
    
    containers = my_html.findAll('div', {'class':'listing-list-item__details__22EsT'}) # get containers
    
    for i in range(len(containers)):
        
        try:
            address = containers[i].findAll('h4', {'class':'SearchListingItem__title__4k6VW listing-list-item__title__map__1U-2y'})[0].text
        except:
            address = 'unknown'
            
        try:
            rooms = containers[i].findAll('div', {'class':'ut_xsHide'})[0].text
        except:
            rooms = 'unknown'
            
        try:
            year = containers[i].findAll('div', {'class':'ut_xsHide'})[1].text
        except:
            year = 'unknown'
            
        try:
            price = containers[i].findAll('div', {'class':'listing-list-item__price__lihQt'})[0].text
        except:
            price = 'unknown'
        
        try:
            bedrooms = containers[i].findAll('span', {'class':'listing-list-item__bedrooms__1kX8p'})[0].text
        except:
            bedrooms = 'unknown'
            
        try:
            bathrooms = containers[i].findAll('span', {'class':'listing-list-item__bathrooms__32Zkv'})[0].text
        except:
            bathrooms = 'unknown'
        
        try:
            pricepsf = containers[i].findAll('div', {'class':'listing-list-item__areaPrice__fcKrS'})[0].text
        except:
            pricepsf = 'unknown'
        
        try:
            sqf = containers[i].findAll('span', {'class':'listing-list-item__sqft__1fFDh'})[0].text
        except:
            sqf = 'unknown'
        
        address_list.append(address)
        rooms_list.append(rooms)
        year_list.append(year)
        price_list.append(price)
        bedrooms_list.append(bedrooms.replace('\xa0',''))
        bathrooms_list.append(bathrooms.replace('\xa0',''))
        pricepsf_list.append(pricepsf.strip())
        sqf_list.append(sqf)
    
    pagenum += 1
    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    try:
        driver.find_element_by_link_text("Next").click()
        print('Page:', pagenum)
    except:
        break

print('End of loop')
```

```python
### consolidate information in dataframe

df = pd.DataFrame({'address': pd.Series(address_list),
                   'rooms': pd.Series(rooms_list),
                   'year': pd.Series(year_list),
                   'price': pd.Series(price_list),
                   'bedrooms': pd.Series(bedrooms_list),
                   'bathrooms': pd.Series(bathrooms_list),
                   'price_psf': pd.Series(pricepsf_list),
                   'sqf_list': pd.Series(sqf_list),})

df.to_csv('property_listings.csv', index=False)
```
